```python
from nltk import WordNetLemmatizer, pos_tag
from nltk.corpus import stopwords, wordnet
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def get_simple_pos(tag) :
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
    
import string
stops = list(set(stopwords.words('english')))
stops += list(string.punctuation)

def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            try:
                pos = pos_tag( [w] ) ### pos is like: [ ('better' , 'RBR') ]
            except IndexError:
                pass
            clean_word = lemmatizer.lemmatize(w , pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
#     Tracer()()

    return output_words
```

```python
x_train_text_words_split = [clean_review(comment.split(' ')) for comment in x_train_text]
```

```python
x_test_text_words_split = [clean_review(comment.split(' ')) for comment in x_test_text]
```
